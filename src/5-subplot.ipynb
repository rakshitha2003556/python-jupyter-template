### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_line = np.arange(0, 10, 0.1)
y_line = np.sin(x_line)

categories = ['A', 'B', 'C']
values = [25, 40, 30]

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

labels = ['Part 1', 'Part 2', 'Part 3', 'Part 4']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Subplot Layout Example')

axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

axs[1, 0].scatter(x_scatter, y_scatter)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

axs[1, 1].pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
axs[1, 1].axis('equal') 
axs[1, 1].set_title('Pie Chart')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
